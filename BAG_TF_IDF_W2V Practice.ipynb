```python
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
import nltk
```

```python
## tokenzation
nltk.download('punkt')
sentance=nltk.sent_tokenize(paragraph)
```

```python
## 
nltk.download('stopwords')
```

```python
stemmer = PorterStemmer()
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
## STEMMING
stemmer.stem('going')
```

```python
## lemmatizer
lemmatizer.lemmatize('going')
```

```python
## text preprocessing
## removing the all special character
## importing the regular expression
import re

corpus = []
for i in range(len(sentance)):
    review = re.sub('[^a-zA-Z]',' ',sentance[i])
    review = review.lower()
    
    corpus.append(review)
```

```python
corpus
```

```python
## applying stemmer on entire document
for i in corpus:
    words = nltk.word_tokenize(i)
    for word in words:
        if word not in set(stopwords.words('english')):
            print(stemmer.stem(word))
```

```python
## applying lemmatizer on entire document
for i in corpus:
    words = nltk.word_tokenize(i)
    for word in words:
        if word not in set(stopwords.words('english')):
            print(lemmatizer.lemmatize(word))
```

```python
## applying stopword on entire document
import re
for i in range(len(sentance)):
    review = re.sub('[^a-zA-Z]',' ',sentance[i])
    review = review.lower()
    review = review.split()
    review = [lemmatizer.lemmatize(word) for word in review if word not in set(stopwords.words('english'))]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
## bag of model
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
bag_count = CountVectorizer(binary=True)
x=bag_count.fit_transform(corpus)
```

```python
x[0].toarray()
```

```python
## TF-IDF MODEL TRAINING
from sklearn.feature_extraction.text import TfidfVectorizer
tf_vec = TfidfVectorizer()
x=tf_vec.fit_transform(corpus)
```
